### Analysis of Launch Site Proximity to the Equator and Coast

To address whether the SpaceX launch sites are near the Equator or the coast, let's examine the latitude of each site and consider their geographical context.

#### 1. Proximity to the Equator
The Equator is located at **0 degrees latitude**. Sites near the Equator would have latitudes between -10 and +10 degrees.

Here are the latitudes for each launch site:
- **CCAFS LC-40**: 28.562302 (far from the Equator)
- **CCAFS SLC-40**: 28.563197 (far from the Equator)
- **KSC LC-39A**: 28.573255 (far from the Equator)
- **VAFB SLC-4E**: 34.632834 (far from the Equator)

Since all of these latitudes are well above +10 degrees, **none of these launch sites are in close proximity to the Equator**.

#### 2. Proximity to the Coast
Launch sites are often located near the coast for safety and logistical reasons, as launching over the ocean reduces risks to populated areas, and coastal locations facilitate transportation of rocket components by sea.

Based on their coordinates:
- **Cape Canaveral Air Force Station (CCAFS)** and **Kennedy Space Center (KSC)** are located near the Atlantic Ocean coast in Florida.
- **Vandenberg Air Force Base (VAFB)** is located close to the Pacific Ocean coast in California.

Thus, **all of these launch sites are in very close proximity to the coast**, allowing for safer launches over the ocean.

## Task 2: Mark the success/failed launches for each site on the map


Next, let's try to enhance the map by adding the launch outcomes for each site, and see which sites have high success rates.
Recall that data frame spacex_df has detailed launch records, and the `class` column indicates if this launch was successful or not


```python
spacex_df.tail(10)
```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
# Create a MarkerCluster object
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python
# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red

# Define a function to assign marker color based on launch outcome
def assign_marker_color(launch_class):
    return 'green' if launch_class == 1 else 'red'

# Apply the function to create the marker_color column
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)

# Verify the new column
spacex_df[['Launch Site', 'Lat', 'Long', 'class', 'marker_color']].tail(10)
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
# For each launch result, add a folium.Marker to the marker_cluster
for index, record in spacex_df.iterrows():
    # Define the coordinates and launch site details
    coordinates = [record['Lat'], record['Long']]
    launch_outcome = 'Success' if record['class'] == 1 else 'Failure'
    site_name = record['Launch Site']
    
    # Create the marker with the appropriate icon color and a popup with site name and outcome
    popup_text = folium.Popup(f"<b>Launch Site:</b> {site_name}<br><b>Outcome:</b> {launch_outcome}", max_width=300)
    marker = folium.Marker(
        location=coordinates,
        icon=folium.Icon(color=record['marker_color']),
        popup=popup_text
    )
    
    # Add the marker to the marker cluster
    marker_cluster.add_child(marker)

# Add the marker cluster to the map
site_map.add_child(marker_cluster)

# Display the map with clustered markers for each launch record
site_map
```

## TASK 3: Calculate the distances between a launch site to its proximities


Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.


You can calculate the distance between two points on the map based on their `Lat` and `Long` values using the following method:

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

### Full Implementation

```python
# Calculate distance and add marker for the closest point of interest
launch_site_lat, launch_site_lon = 28.56341, -80.57678  # Launch site coordinates
coastline_lat, coastline_lon = 28.56332, -80.56755      # Closest coastline coordinates

# Calculate the distance to the closest coastline
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Add a marker on the map at the coastline with the distance displayed
distance_marker = folium.Marker(
    location=[coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
    )
)

# Add the distance marker to the site map
site_map.add_child(distance_marker)

# Display the updated map
site_map
```

```python
# Points of interest around a specific launch site
points_of_interest = {
    'Closest City1': [28.6122, -80.8076],  # City coordinates: Titusville, FL
    'Closest City2': [28.3922, -80.6077],  # City coordinates: Cape Canaveral, FL
    'Railway': [28.5721, -80.5853],        # Railway coordinates
    'Highway': [28.56366, -80.57081]       # Highway coordinates
}

# Define the launch site coordinates (replace with the actual launch site coordinates)
launch_site_coords = [28.56341, -80.57678]

# Iterate over each point of interest, calculate the distance, and add markers and lines
for point_name, point_coords in points_of_interest.items():
    # Calculate the distance between the launch site and the point of interest
    distance = calculate_distance(launch_site_coords[0], launch_site_coords[1], point_coords[0], point_coords[1])
    
    # Add a marker at the point of interest with the distance in the popup
    distance_marker = folium.Marker(
        location=point_coords,
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12; color:#d35400;"><b>{:.2f} KM</b></div>'.format(distance),
        ),
        popup=f"{point_name} ({distance:.2f} KM)"
    )
    site_map.add_child(distance_marker)
    
    # Draw a PolyLine from the launch site to the point of interest
    line = folium.PolyLine(locations=[launch_site_coords, point_coords], weight=1, color='blue')
    site_map.add_child(line)

# Display the map with markers and lines
site_map
```
